```python
# Задание №7.Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом
salary_low = df[df['salary'] == 'low']
salary_high = df[df['salary'] == 'high']
salary_low= salary_low.sample (n= 1237 )
```

```python
salary_low_av = salary_low['average_montly_hours']
salary_high_av = salary_high['average_montly_hours']
```

```python
#Гипотеза
#H0: Средние выборок одинаковы  / сотрудники с высоким окладом проводят на работе столько же времени, чем сотрудники с низким окладом
#H1: Средние выборок различаются / сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом
```

```python
stat, p = st.ttest_ind(salary_low_av, salary_high_av)

print(f"Статистика = {stat:.5f}, p = {p:.10f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, средние, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, средние, вероятно, различаются')
```

```python
# Задание №9.Разделить данные на тестовую и обучающую выборки
# Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary)
# Оценить качество модели на тестовой выборки
```

```python
data = df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'left', 'promotion_last_5years']]
#убираем столбцы department, salary и left
data_class= df['left'] # классы -уволился / не уволился
data
```

```python
# разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, data_class, random_state=42)
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем данные
lda.fit(X_train, y_train)
```

```python
X_test.reset_index(drop= True , inplace= True )
y_test.reset_index(drop= True , inplace= True )
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
result
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(X_test))
```
